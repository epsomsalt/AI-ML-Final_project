```python
import openai
import os

openai.api_key = "Your_openai_api_key"

def generate_text(prompt, model):
    completions = openai.Completion.create(
        engine=model,
        prompt=prompt,
        max_tokens=1000,
        n=1,
        stop=None,
        temperature=0.5,
    )

    message = completions.choices[0].text
    return message

input_text = "NLP is natural language processing"
model_name = "davinci"

output_text = generate_text(input_text, model_name)

print("Input text: ", input_text,)
print("Output text: ", input_text+output_text)
```
